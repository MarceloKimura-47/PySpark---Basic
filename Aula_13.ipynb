```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = (
    SparkSession.builder.appName('Curso PySpark')
    .config('spark.sql.repl.eagerEval.enabled',True)
    .getOrCreate()
)

spark
```

```python
df = spark.read.parquet('C:/Users/Kimura/Desktop/Curso PySpark/DATASETS/DATASETS/LIVROS.parquet').limit(5)
df
```

```python
# O união funciona como se fosse um empilhamento de DataFrames. O requisito para esse método funcionar é que as colunas sejam iguais nos 2 DataFrames que serão conectados.

dup = df.union(df)
dup
```

```python
df1 = df.select('cnpj_editora','data_lancamento', 'ean', 'isbn10', 'numero_paginas','preco','id')
```

```python
# Veja que o método funcionou, porém como os dados estão fora de ordem isso gerou essa configuração errada na tabela.
# Caso a quantidade de colunas nos dois DataFrames sejam diferentes, o código não funciona!

df.union(df1)
```

```python
# Esse método procura as colunas correspondentes e faz o empilhamento dos dados.

df.unionByName(df1)
```

```python
df2 = df1.withColumn('qtd', F.lit(1))
df2
```

```python
# Nesse caso temos 2 DataFrames semelhantes, mas 1 deles tem 1 coluna a mais, nesse caso podemos realizar essa associação usando o parâmetro allowMissingColumns=True
# Para a correspondencia não existente o valor é nulo.

df2.unionByName(df1, allowMissingColumns=True)
```
